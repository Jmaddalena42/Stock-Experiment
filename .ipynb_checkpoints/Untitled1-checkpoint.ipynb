```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import talib as ta
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('AAPL.csv')
#df.index = df['Date']
df.head()
```

```python
#Exponential Moving Average  
ema = ta.EMA(df.Close,15)
df['MA15'] = ema
df = df.fillna(0)
df.tail()
```

```python
df2 = df[['Date', 'Close', 'MA15']].copy()

df2.index = df['Date']

df2 = df2.drop(["Date"], axis=1)


df2.tail()
```

```python
N = 5
df['Uptrend'] = ''
for index,row in df.iterrows():
    if (df.loc[index-N:index,"EMA"] > df.loc[index-N:index,"EMA"].shift(-1)):
        df.loc[index, 'Uptrend'] = "Uptrend"
    elif (df.loc[index-N:index,"EMA"] < df.loc[index-N:index,"EMA"].shift(-1)):
        df.loc[index, 'Uptrend'] = "Downtrend"
    elif (df.loc[index-N:index,"EMA"] == 0):
        df.loc[index, 'Uptrend'] = "Notrend"  
    else:
        df.loc[index, 'Uptrend'] = "Notrend"
```

```python
if df['close'] > df['EMA'] and df['EMA'] > df['EMA'].shift(-1) and df['EMA'].shift(-1) > df['EMA'].shift(-2) and df['EMA'].shift(-2) > df['EMA'].shift(-3):
    df['trend'] = 1
else:
    df['trend'] = 0
```
